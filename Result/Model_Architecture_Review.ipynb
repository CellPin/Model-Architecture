### 1. PaperModel(No Freeze)
* PaperModel을 처음 사용할 당시 우리가 Fine Tunning에 대해 잘못 알고 있어서 pre_trained 된 부분을 Freeze 하지 않고, 분류기를 합친 뒤 모두 학습을 시켰다.
* 그리고 나중에 pre_trained 된 부분은 Freeze을 시키고, 새로운 분류기만 학습을 시킨 뒤 Fine Tunning 하는 것이라는 것을 알아서 PaperModel을 다시 Fine Tunning 하였다.
* 하지만 Freeze을 하지 않은 모델이 loss와 accuracy가 좋아서 모델 비교 후보에서 제외시키지 않았다.
